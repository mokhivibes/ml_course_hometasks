```python
#Vazifa 1 — Array yaratish
'''np.zeros, np.ones, np.arange ishlatib quyidagilarni yarating:
5 ta noldan iborat array
3×3 birlardan iborat array
2 dan 20 gacha toq sonlar (qadam 2)
Har birining shape va dtype ni chiqaring.'''

import numpy as np

array = np.zeros(5)
print(array)
print(array.shape)
print(type(array))

array2 = np.ones((3, 3))
print(array2)
print(array2.shape)
print(type(array2))

toq_sonlar = np.arange(3, 20, 2)
print(toq_sonlar)
print(toq_sonlar.shape)
print(type(toq_sonlar))
```

```python
#Vazifa 2 — Array xususiyatlari
'''Quyidagi arrayni yarating:
arr = np.array([[4, 7, 2], [1, 9, 5], [8, 3, 6]])
shape, ndim, size, dtype ni chiqaring.
Natijalarni o'zingiz tushuntiring — har biri nima degani?'''

arr = np.array([[4, 7, 2],
                [1, 9, 5],
                [8, 3, 6]])
print("Shape:" , arr.shape, ", Number of dimension:", arr.ndim, ", Size:", arr.size, ", Type:", arr.dtype)
```

```python
#Vazifa 3 — Indexing va Slicing
'''Yuqoridagi arr dan:
Birinchi qatorni oling
Ikkinchi ustunni oling
9 elementini indeks bilan oling
Yuqori o'ng 2×2 qismni oling'''

arr = np.array([[4, 7, 2],
                [1, 9, 5],
                [8, 3, 6]])
qator1 = arr[0]
print(qator1)

ustun2 = arr[0:3, 1]
print(ustun2)

element9 = arr[1, 1]
print(element9)

qism2x2 = arr[0:2, 1:3]
print(qism2x2)
```

```python
#Vazifa 4 — Random
'''np.random.seed(42) ishlatib:
1 dan 100 gacha 10 ta tasodifiy butun son yarating
0 dan 1 gacha 3×3 float array yarating
Har ikkalasining mean va max ini chiqaring.'''

np.random.seed(42)
arr = np.random.randint(1, 100, 10)
print(arr)
print("Mean:", arr.mean(), "Max:", arr.max())

arr2 = np.random.rand(3, 3)
print(arr2)
print("Mean:",arr2.mean(), "Max:", arr2.max())
```

```python
#Vazifa 5 — 3D array
'''np.ones ishlatib 2×3×4 o'lchamli array yarating.
ndim, shape, size ni chiqaring va nima ekanligini tushuntiring
Birinchi 'qatlamdan' birinchi qatorni oling'''

array = np.ones((2,3,4))
print(array)
print(array.ndim, array.shape, array.size)
print(array[0, 0])
```
