```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
sns.set_theme()
```

Vazifa 1 — Oddiy Line Plot [10, 20, 30, 40, 50] x va [5, 15, 10, 25, 20] y qiymatlari uchun grafik chizing. Chiziq ko'k rangda, marker o shaklida bo'lsin. Sarlavha va o'q nomlari qo'shing.


```python
x = [10, 20, 30, 40, 50]
y = [5, 15, 10, 25, 20]
fig, ax = plt.subplots()
ax.plot(x, y,
        color = "blue",
        marker = "o")
ax.set_xlabel("X axe")
ax.set_ylabel("Y axe")
ax.set_title('Line Graph')
plt.show()
```

Vazifa 2 — Bar Chart 5 ta o'quvchining baholari: Ali: 88, Vali: 72, Sardor: 95, Kamola: 60, Zulfiya: 81. Vertikal bar chart chizing. Eng yuqori ball ustuni boshqa rangda ajralib tursin.


```python
students = ["Ali", "Vali", "Sardor", "Kamola", "Zulfiya"]
grades = np.array([88, 72, 95, 60, 81])
colors = ['blue', 'blue', 'blue', 'blue', 'blue']
max_grade = grades.argmax()
colors[max_grade] = 'red'
fig, ax = plt.subplots()
ax.bar(students, grades,
        color = colors
)
ax.set_xlabel("Students")
ax.set_ylabel("Grades")
ax.set_title('Bar Graph')

plt.show()

```

Vazifa 3 — Gorizontal Bar Chart Quyidagi dasturlash tillari mashhurligini ko'rsating: Python: 90, JavaScript: 85, Java: 70, C++: 65, R: 40. Ustunlar bir xil rangda bo'lmasin, o'q nomlari va sarlavha bo'lsin.

```python
import random
fig, ax = plt.subplots()
languages = ["Python", "JavaScript", "Java", "C++", "R"]
popularity = np.array([90, 85, 70, 65, 40])
colors = ['green', 'yellow', 'red', 'black', 'pink']
ax.barh(languages, popularity,
        color = colors)

ax.set_title("Programming Language's Popularity")
ax.set_ylabel("Languages")
ax.set_xlabel("Popularity in %")
```

Vazifa 4 — Scatter Plot Numpy bilan 100 ta tasodifiy x va y qiymatlari generatsiya qiling. Scatter plot chizing, alpha 0.5, nuqta rangi yashil bo'lsin. Sarlavhada "Tasodifiy nuqtalar" deb yozing.


```python
x = np.random.randint(0, 100, 100)
y = 3.8 * x + np.random.normal(0, 20, 100)
fig, ax = plt.subplots()
ax.scatter(x, y,
           alpha = 0.5,
           color = "green")

ax.set_title("Tasodifiy nuqtalar")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

Vazifa 5 — Histogram np.random.normal(loc=170, scale=10, size=300) — 300 ta bo'y qiymati. Histogramma chizing, bins=25, ustunlar orasida oq chegara bo'lsin. X o'qi "Bo'y (sm)", Y o'qi "Soni" bo'lsin.


```python
data = np.random.normal(loc=170, scale=10, size=300)

fig, ax = plt.subplots()
ax.hist(data, bins=25, edgecolor="white")

ax.set_xlabel("Bo'y (sm)")
ax.set_ylabel("Soni")
ax.set_title("Bo'y taqsimoti")

plt.show()
```

Vazifa 6 — Pie Chart Bir sinfdagi qiziqishlar: Sport: 30%, Musiqa: 25%, Kitob: 20%, Gaming: 15%, Boshqa: 10%. Pie chart chizing. Eng katta qism explode bilan ajralib tursin, foizlar ko'rinsin.


```python
qiziqishlar = ["Sport", "Musiqa", "Kitob", "Gaming", "Boshqa"]
foizlar = [30, 25, 20, 15, 10]
fig, ax = plt.subplots()
ax.pie(
    foizlar,
    labels = qiziqishlar,
    explode = [0.3, 0.1, 0.1, 0.1, 0.1 ],
    autopct = "%1.1f%%"
)
plt.show()
```

Vazifa 7 — Subplots 2x2 Bitta Figure ichida 4 ta grafik:
Yuqori chap: Line plot
Yuqori o'ng: Bar chart
Quyi chap: Scatter plot
Quyi o'ng: Histogram
Har birida sarlavha bo'lsin, tight_layout ishlatish shart.


```python
fig, ax = plt.subplots(2, 2, figsize = (10, 10))

#Linegraph
x = [10, 20, 30, 40, 50]
y = [5, 15, 10, 25, 20]

ax[0,0].plot(x, y,
        color = "blue",
        marker = "o")
ax[0,0].set_xlabel("X axe")
ax[0,0].set_ylabel("Y axe")
ax[0,0].set_title('Line Graph')

#Bargraph
students = ["Ali", "Vali", "Sardor", "Kamola", "Zulfiya"]
grades = np.array([88, 72, 95, 60, 81])
colors = ['blue', 'blue', 'blue', 'blue', 'blue']
max_grade = grades.argmax()
colors[max_grade] = 'red'

ax[0,1].bar(students, grades,
        color = colors
)
ax[0,1].set_xlabel("Students")
ax[0,1].set_ylabel("Grades")
ax[0,1].set_title('Bar Graph')

#Scatter Plot
x = np.random.randint(0, 100, 100)
y = 3.8 * x + np.random.normal(0, 20, 100)

ax[1,0].scatter(x, y,
           alpha = 0.5,
           color = "green")

ax[1,0].set_title("Tasodifiy nuqtalar")
ax[1,0].set_xlabel("x")
ax[1,0].set_ylabel("y")

#Histogram
data = np.random.normal(loc=170, scale=10, size=300)


ax[1,1].hist(data, bins=25, edgecolor="white")
ax[1,1].set_xlabel("Bo'y (sm)")
ax[1,1].set_ylabel("Soni")
ax[1,1].set_title("Bo'y taqsimoti")

plt.tight_layout()
plt.show()
```

Vazifa 8 — Grafik bezash Istalgan line plot chizing. Quyidagilarning barchasini qo'shing: grid, legend, chiziq rangi va linestyle, marker, sarlavha, o'q nomlari, figsize=(10, 5).


```python
fig, ax = plt.subplots(figsize=(10, 5))
x = [10, 15, 20, 28, 35]
y = [5, 20, 15, 25, 18]
ax.plot(x, y,
        color = "#126263",
        marker = "s",
        label ="Just a Linegraph",
        linestyle = ":"
        )
ax.set_xlabel("X axe")
ax.set_ylabel("Y axe")
ax.set_title('Line Graph')
ax.legend()
ax.grid()
plt.show()
```

Vazifa 9 — Ko'p chiziqli grafik Bitta grafikda 3 ta o'quvchining 5 ta fandagi ballari ko'rsatilsin. Har bir o'quvchi boshqa rangda va boshqa markerde bo'lsin. Legend orqali kim ekanini aniq ko'rsating.


```python
import pandas as pd
subjects = ['Math', 'English', 'Science', 'Geography', 'History']
Aziza = [78, 85, 64, 97, 45]
Rustam = [87, 65, 77, 33, 59]
Timur = [65, 72, 90, 51, 98]
fig, ax = plt.subplots()
ax.plot(subjects, Aziza, color = "blue", marker = "o", label = "Aziza")
ax.plot(subjects, Rustam, color = "red", marker = "s", label = "Rustam")
ax.plot(subjects, Timur, color = "green", marker = "D", label = "Timur")
ax.legend()
ax.set_title("Students' Grades")
ax.set_xlabel("Subjects")
ax.set_ylabel("Grades")
plt.show()
```

```python
import pandas as pd
subjects = ['Math', 'English', 'Science', 'Geography', 'History']
Aziza = [78, 85, 64, 97, 45]
Rustam = [87, 65, 77, 33, 59]
Timur = [65, 72, 90, 51, 98]
fig, ax = plt.subplots()
ax.plot(subjects, Aziza, color = "blue", marker = "o", label = "Aziza")
ax.plot(subjects, Rustam, color = "red", marker = "s", label = "Rustam")
ax.plot(subjects, Timur, color = "green", marker = "D", label = "Timur")
ax.legend()
ax.set_title("Students' Grades")
ax.set_xlabel("Subjects")
ax.set_ylabel("Grades")
plt.savefig("students_result.png", dpi=200, bbox_inches="tight")
plt.show()

```
