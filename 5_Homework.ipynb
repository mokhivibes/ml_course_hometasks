🐧 Uyga vazifa — Dars 5: Seaborn
Dataset: penguins (sns.load_dataset("penguins"))


Topshiriq 1 — Muhitni sozlash
sns.set_theme() yordamida quyidagilarni o'rnating:

-> style="whitegrid", context="notebook", palette="muted"

-> Datasetni yuklab, df.head() va df.info() ni chaqiring


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df = sns.load_dataset("penguins")
```

```python
sns.set_theme(
    style="whitegrid",
    context="notebook",
    palette="muted",
)
print(df.head())
print(df.info())
```

Topshiriq 2 — Histogramma
bill_length_mm ustuni bo'yicha histogramma chizing:

->bins=30, kde=True

->Sarlavha: "Tumshug' uzunligi taqsimoti"


```python
sns.histplot(data=df, x="bill_length_mm", bins=30, kde=True)
plt.title("Tumshug' uzunligi taqsimoti")
plt.tight_layout()
plt.show()
```

Topshiriq 3 — Hue bilan histogramma
Topshiriq 2 ni takrorlang, lekin hue="species" qo'shing.
 Qaysi penguin turi tumshug'i uzunroq? — 1 gap izoh yozing


```python
sns.histplot(data=df, x="bill_length_mm", bins=30, kde=True, hue="species")
plt.title("Tumshug' uzunligi taqsimoti")
plt.tight_layout()
plt.show()
print("\nEng tumshug'i uzun penguin turi bu Chinstrap.")
```

Topshiriq 4 — KDE grafik
body_mass_g uchun kdeplot chizing:
hue="species", fill=True, alpha=0.4
Uch turning og'irligi taqsimoti bir-biridan farq qiladimi? — izohlang


```python
sns.kdeplot(data=df, x="body_mass_g", hue="species", fill=True, alpha=0.4)
plt.title("Tana og'irligi taqsimoti — penguin turlari bo'yicha")
plt.tight_layout()
plt.show()
print("\nAdelie va Chinstrap og'irligi bir biriga yaqin lekin Gentoo og'irligi qolgan 2 turga qaraganda og'irroq ya'ni farq qiladi.")
```

Topshiriq 5 — Boxplot
species bo'yicha flipper_length_mm uchun boxplot chizing:

-> hue="sex" qo'shing

-> Eng katta IQR qaysi turda? — komentda yozing


```python
sns.boxplot(data=df, x="species", y="flipper_length_mm", hue="sex")
plt.title("Qanotlar uzunligi — tur va jins bo'yicha")
plt.tight_layout()
plt.show()
print("\nEng katta IQR — Gentoo turida (Q3-Q1 farqi eng katta=> male=225~218, female=215~210)")
```

Topshiriq 6 — Violinplot
species bo'yicha bill_depth_mm uchun violinplot chizing:

-> inner="box", hue="sex"

-> Boxplot bilan solishtirganda qanday qo'shimcha ma'lumot ko'rindi? — izohlang


```python
sns.violinplot(data=df, x="species", y="bill_depth_mm", inner="box", hue="sex")
plt.title("Tumshug' chuqurligi — tur va jins bo'yicha")
plt.tight_layout()
plt.show()
print("\nViolinplot da qiymatlar taqsimoti ham ko'rsatiladi")
```

Topshiriq 7 — Stripplot
island bo'yicha body_mass_g uchun stripplot chizing:

-> hue="species", jitter=True, alpha=0.6

-> Qaysi orolda eng og'ir penguinlar yashaydi? — komentda yozing


```python
sns.stripplot(data=df, x="island", y="body_mass_g", hue="species", jitter=True, alpha=0.6)
plt.title("Orol bo'yicha tana og'irligi")
plt.tight_layout()
plt.show()
#Biscoe island da eng o'gir penguinlar yashaydi
```

Topshiriq 8 — Regplot
bill_length_mm (x) va bill_depth_mm (y) orasidagi munosabat uchun regplot chizing:

-> ci=95

-> Tumshug' uzunligi va chuqurligi o'rtasida bog'liqlik bormi? — 1-2 gap izohlang


```python
sns.regplot(data=df, x="bill_length_mm", y="bill_depth_mm", ci=95)
plt.title("Tumshug' uzunligi va chuqurligi o'rtasida bog'liqlik")
plt.tight_layout()
plt.show()
print("\nTumshug' uzunligi ortgani sari, tumshug' chuqurligi kamayadi, manfiy bog'liqlik")
```

Topshiriq 9 — Jointplot
flipper_length_mm va body_mass_g uchun jointplot chizing:

-> kind="reg", hue="species"

-> Qanotlari uzunroq penguinlar og'irroqmi? — xulosa yozing


```python
sns.jointplot(data=df, x="flipper_length_mm", y="body_mass_g", kind="reg")
plt.show() #hue kind=reg da ishlamas ekan, shuning uchu pastda kind=scatter bilan grafni boshqatta yaratib ko'ramiz

sns.jointplot(data=df, x="flipper_length_mm", y="body_mass_g", hue="species", kind="scatter")
plt.title("Qanotcha vs og'irlik (regression)")
plt.tight_layout()
plt.show()
#Qanotlari uzunroq penguinlar og'irroq, ya'ni Gentoolar"
```

Topshiriq 10 — Subplots (umumiy)
fig, axes = plt.subplots(1, 3, figsize=(15, 5)) ichida:

-> Chap: species bo'yicha body_mass_g — boxplot

-> O'rta: species bo'yicha flipper_length_mm — violinplot

-> O'ng: species bo'yicha bill_length_mm — stripplot


```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
sns.boxplot(data=df, x="species", y="body_mass_g", ax=axes[0])
sns.violinplot(data=df, x="species", y="flipper_length_mm", ax=axes[1])
sns.stripplot(data=df, x="species", y="bill_length_mm", ax=axes[2])
plt.title("Tana og'irligi, qanot uzunligi, tumshug' uzunligi bo'yicha turlar bo'yicha taqsimot")
plt.tight_layout()
plt.show()
```
