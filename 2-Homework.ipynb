```python
#5-mashq
'''Vazifa
3-daraja — List bilan ishlash
Sinf natijalarini tahlil qilish
Real vaziyat
O‘qituvchi sinf natijalarini tezda tahlil qilmoqchi.
ballar = [72, 88, 95, 61, 79, 85]
Vazifa
Berilgan ballar bo‘yicha:
Eng yuqori ball
Eng past ball
O‘rtacha ball
ni hisoblang.'''

ballar = [72, 88, 95, 61, 79, 85]
def find_results(balls):
  maximum = max(balls)
  print(f"Eng yuqori ball: {maximum}")
  minimum = min(balls)
  print(f"Eng past ball: {minimum}")
  medium = sum(balls) // len(balls)
  print(f"O'rtacha ball: {medium}")

find_results(ballar)

#########################
```
